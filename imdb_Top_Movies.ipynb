```python
def duration_to_minutes(duration_str):
    hours = 0
    minutes = 0

    if 'h' in duration_str:
        hours = int(duration_str.split('h')[0])
        if 'm' in duration_str:
            minutes = int(duration_str.split('h')[1].split('m')[0])
        else:
            minutes = 0
    elif 'm' in duration_str:
        minutes = int(duration_str.split('m')[0])
        hours = 0

    return hours * 60 + minutes

movies['Duration To Minutes'] = movies['Duration'].apply(duration_to_minutes)
movies_duration_minutes = movies[['Title', 'Duration To Minutes']]
movies_duration_minutes
```

#GROUPED DATA ANALYSIS


```python
yearly_rating = movies.groupby('Year')['Rating'].mean()
print(yearly_rating)
```

```python
rating_counts = movies.groupby('Rating').size().reset_index(name='Count')
rating_counts
```

```python
yearly_counts = movies.groupby('Year').size().reset_index(name='Count')
yearly_counts
```

```python
yearly_rating_counts = movies.groupby(['Year', 'Rating']).size().reset_index(name='Count')
yearly_rating_counts
```

```python
yearly_certificate_counts = movies.groupby(['Year', 'Certificate']).size().reset_index(name='Count')
yearly_certificate_counts
```

```python
genre_yearly_avg_rating = movies.groupby(['Year', 'Genres'])['Rating'].mean().reset_index()
genre_yearly_avg_rating
```

```python
pivot_table_rating_stats = movies.pivot_table(index='Year', values='Rating', aggfunc=['mean', 'min', 'max', 'std']).fillna('--')
pivot_table_rating_stats
```

```python
pivot_table_certificate_year = movies.pivot_table(index='Year', columns='Certificate', values='Title', aggfunc='count').fillna(0)
print(pivot_table_certificate_year)
```

```python
pivot_table_avg_duration_genre = movies.pivot_table(index='Genres', values='Duration To Minutes', aggfunc='mean').fillna(0)
pivot_table_avg_duration_genre
```

```python
pivot_table_max_rating_cert_genre = movies.pivot_table(index='Certificate', columns='Genres', values='Rating', aggfunc='max').fillna(0)
pivot_table_max_rating_cert_genre
```
